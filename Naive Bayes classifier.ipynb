```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/wheat-seeds.csv"
```

```python
headernames = ['seed 1','seed 2','seed 3','seed 4','seed 5','seed 6','seed 7','class']
```

```python
data = pd.read_csv(path,names = headernames)
```

let's Implement the basic step of data preprocessing

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

Here we are dividing the data into train and test splits, with 70% of the training data and 30% of testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20)
```

let's do the data scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

Now we are training the model with the help of the GaussianNB class of sklearn as follows

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
```

Now, we have to make the prediction

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
```

```python
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```
